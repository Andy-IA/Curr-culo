```python
from sklearn.datasets import load_iris
import pandas as pd

# Carregar o dataset
iris = load_iris()
dados = pd.DataFrame(iris.data, columns=iris.feature_names)
dados['target'] = iris.target  # Adicionar a coluna de classificação
```

```python
# Primeiras linhas
dados.head()

# Estatísticas descritivas
dados.describe()
```

```python
from sklearn.model_selection import train_test_split

# Separar features (X) e target (y)
X = dados.drop('target', axis=1)
y = dados['target']

# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Criar o modelo
modelo = RandomForestClassifier(n_estimators=100, random_state=42)

# Treinar o modelo
modelo.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Fazer previsões
y_pred = modelo.predict(X_test)

# Calcular a acurácia
acuracia = accuracy_score(y_test, y_pred)
print(f'Acurácia do modelo: {acuracia:.2f}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gráfico de dispersão entre comprimento e largura das pétalas
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target', data=dados, palette='viridis')
plt.title('Gráfico de Dispersão: Comprimento vs Largura das Pétalas')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Calcular a matriz de confusão
matriz_confusao = confusion_matrix(y_test, y_pred)

# Plotar a matriz de confusão
sns.heatmap(matriz_confusao, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Matriz de Confusão')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```

```python
# Extrair a importância das features
importancia = modelo.feature_importances_

# Plotar a importância
sns.barplot(x=importancia, y=iris.feature_names, palette='viridis')
plt.title('Importância das Features')
plt.show()
```
